```python
import pandas as pd
import geopandas as gpd
```

```python
# read the geopackage
gdf = gpd.read_file("SIRGAS_GPKG_estacaometro_com_inaug.gpkg")
gdf
```

```python
# the "date" column is called "data_inaug".
# we have a list of years:
years = list(range(1995, 2020))
years
```

```python
# I want to filter the gdf to create a new gdf to each date. Only the rows where the date is lower than the year must be kept
dictionary = {}
# Iterate over each year
for year in years:
    # Convert 'data_inaug' to datetime, considering that the format is day-month-year
    if gdf['data_inaug'].dtype == 'object':
        gdf['data_inaug'] = pd.to_datetime(gdf['data_inaug'], dayfirst=True)

    # Filter the GeoDataFrame to only include rows where the year of 'data_inaug' is less than the current year
    gdf_filtered = gdf[gdf['data_inaug'].dt.year < year]
    dictionary[year] = gdf_filtered
```

```python
import os

# Define the folder where you want to save the files
output_folder = 'output_geopackages'

# Create the folder if it does not exist
os.makedirs(output_folder, exist_ok=True)

# Iterate over the dictionary and save each GeoDataFrame as a separate GeoPackage
for year, gdf_filtered in dictionary.items():
    # Define the file name based on the year
    file_name = f"eatacoes_metro_ate_{year}.gpkg"

    # Define the full path for the file
    file_path = os.path.join(output_folder, file_name)

    # Save the GeoDataFrame as a GeoPackage
    gdf_filtered.to_file(file_path, driver='GPKG')

print("All GeoPackage files have been saved.")
```
