```python
import pandas as pd
```

```python
df = pd.read_csv("ipca_history.csv")
df.head()
```

```python
# Calculando a inflação acumulada anual
df["ipca_ano"] = df.iloc[:, 1:].apply(lambda x: (x / 100 + 1).prod(), axis=1) - 1

# Convertendo para porcentagem
# df['ipca_ano'] *= 100

# roud to 4 decimal places
df["ipca_ano"] = df["ipca_ano"].apply(lambda x: round(x, 4))

# Exibindo os resultados
df[["Ano", "ipca_ano"]]
```

## Calcular a inflação acumulada até o ano de 2023 para cada ano


```python
# Convertendo as taxas para forma decimal
# df['ipca_ano'] /= 100

# Calculando a inflação composta de cada ano até 2023
ano_final = 2023
df["ipca_ate_2023"] = df.apply(
    lambda row: df[(df["Ano"] >= row["Ano"]) & (df["Ano"] <= ano_final)]["ipca_ano"]
    .add(1)
    .prod()
    - 1,
    axis=1,
)

# Convertendo a taxa composta de volta para porcentagem
# df['ipca_ate_2023'] *= 100

# round to 4 decimal places
df["ipca_ate_2023"] = df["ipca_ate_2023"].round(4)

# visualize the data
df[["Ano", "ipca_ate_2023"]]
```

```python
# Filter columns
df = df[["Ano", "ipca_ano", "ipca_ate_2023"]]

# save to csv file
df.to_csv("ipca_history_yearly.csv", index=False)
```
