```python
import pandas as pd
import geopandas as gpd
```

```python
# importar o dataset tabela 1 das zonas origem e destino,
df1 = pd.read_excel("../OD-2017/OD-2017/Tabelas-OD2017/Tab01_OD2017.xlsx", skiprows=7, skipfooter=1)
```

```python
# mantenha somente as linhas em que a coluna "Zona" possa ser convertida para inteiro
df1 = df1[df1["Zona "].astype(str).str.isdigit()]

# removar colunas Produzidas e Atraidas
df1 = df1.drop(columns=["Produzidas", "Atraidas"])
df1.head()
```

```python
zonas_cidade_sp = gpd.read_file("zonas_od_sp_corrected.gpkg")
zonas_cidade_sp
```

```python
# pegar uma lista de valores unicos da coluna "od_id" do zonas_cidade_sp
zonas_cidade_sp_ids = zonas_cidade_sp["od_id"].unique()
# converter tudo para inteiros
zonas_cidade_sp_ids = zonas_cidade_sp_ids.astype(int)
zonas_cidade_sp_ids
```

```python
# Vamos aplicar um filtro aqui: manter somente as linhas do df1 em que a coluna "Zona " esteja contida na lista zonas_cidade_sp_ids
df1 = df1[df1["Zona "].isin(zonas_cidade_sp_ids)]
df1.head()
```

```python
# ler a tabela 16 da matriz de origem e destino  
df06 = pd.read_excel("../OD-2017/OD-2017/Tabelas-OD2017/Tab06_OD2017.xlsx", skiprows=7, skipfooter=1)
```

## Fazer join da tabela 06

```python
# Manter somente as linhas em que a coluna Residência esteja dentro da lista zonas_cidade_sp_ids
df06 = df06[df06["Residência"].isin(zonas_cidade_sp_ids)]

# remover colunas que começam com o nome "Unnamed"
df06 = df06.loc[:, ~df06.columns.str.startswith('Unnamed')]

# manter somete as colunas "Redidência" e "Per Capita"
df06 = df06[["Residência", "Per Capita"]]
```

```python
df06
```

```python
#fazer o join em uma tabela só: unir df1 e df06, com as chaves sendo a coluna "Zona " do df1 e a coluna "Residência" do df06
df_od = df1.merge(df06, left_on="Zona ", right_on="Residência")
# remover a coluna "Residência" do df_od
df_od = df_od.drop(columns=["Residência"])
```

```python
df_od
```

## normalizar os dados entre 0 e 1

```python
names = ["Zona ", "Nome"]  
norm_columns = [col for col in df_od.columns if col not in names]
norm_columns
```

```python
# normalizar todas as colunas para terem valores entre 0 e 1
df_od_norm = df_od.copy()
for col in norm_columns:
    df_od_norm[col] = df_od_norm[col] / df_od_norm[col].max()
```

```python
df_od_norm
```

## remover variaveis que são muito parecidas umas das outras

```python
# gerar uma matriz de coeficiente de correlação entre as colunas que estejam na lista norm_columns
# meu objetivo é checar se alguma coluna é muito correlacionada com outra, pois isso pode causar problemas na hora de rodar o algoritmo
df_od_norm[norm_columns].corr()
```

```python
# aponte-me colunas que tenham correlação maior que 0.8 ou menor que -0.8
list_names = [x for x in df_od_norm.columns if x not in names]
matrix = df_od_norm[list_names].corr().abs() > 0.8

# agora imprima os pares em que a correlação é maior que 0.8 ou menor que -0.8, ou seja, a matriz tenha True na célula
matrix[matrix == True].stack() 
```
